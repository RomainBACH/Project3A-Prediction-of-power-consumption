```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
consumptionTetuan=pd.read_csv("Tetuan City power consumption.csv")
```

```python
zone1 = consumptionTetuan['Zone 1 Power Consumption']
date = consumptionTetuan['DateTime']
```

```python
features=['Temperature', 'Humidity', 'Wind Speed', "general diffuse flows", "diffuse flows"]
X = consumptionTetuan[features]
y = zone1
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train
```

```python
X_test
```

```python
model_KNN = KNeighborsRegressor(n_neighbors=5)


model_KNN.fit(X_train, y_train)
print(model_KNN.score(X_train, y_train)) # R^2
predictions_KNN = model_KNN.predict(X_test)
print("MAE =", mean_absolute_error(y_test, predictions_KNN))
```

```python
model_TreeDecision = DecisionTreeRegressor(random_state=1)

model_TreeDecision.fit(X_train, y_train)
print(model_TreeDecision.score(X_train, y_train))
predictions_TreeDecision= model_TreeDecision.predict(X_test)
print("MAE =", mean_absolute_error(y_test, predictions_TreeDecision))
```

```python
forest_model = RandomForestRegressor(random_state=1)
forest_model.fit(X_train, y_train)
print(forest_model.score(X_train, y_train))
predictions_Forest = forest_model.predict(X_test)
print("MAE =", mean_absolute_error(y_test, predictions_Forest))
```

```python
print("Size of y_test:",y_test.shape)
print("Size of y_train:", y_train.shape)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
mean = X_train.mean()
std  = X_train.std()
X_train_norm = (X_train - mean) / std
X_test_norm = (X_test - mean) / std
```

```python
shape=X_train_norm.shape[1] 

model = keras.models.Sequential()
model.add(keras.layers.Input(shape , name="InputLayer")) 
model.add(keras.layers.Dense(6, activation='relu', name='Dense_n1'))
model.add(keras.layers.Dense(12, activation='relu', name='Dense_n2'))
model.add(keras.layers.Dense(1, name='Output'))

model.compile(optimizer = 'adam',
              loss      = 'mae',
              metrics   = 'mae' )

model.summary()
```

```python
epochs = 60

history = model.fit(X_train_norm,
                    y_train,
                    epochs          = epochs,
                    batch_size      = 100,
                    verbose         =  1,
                    validation_data = (X_test_norm, y_test))
```

```python
predictions = model.predict(X_test_norm)
```

```python
print("MAE =", mean_absolute_error(y_test, predictions))
```

```python
X_test['index']= range(1,10485)

X_test['Zone 1'] = y_test
X_test['Zone 1 predictions KNN'] = predictions_KNN
X_test['Zone 1 predictions Tree'] = predictions_TreeDecision
X_test['Zone 1 predictions Forest'] = predictions_Forest
X_test['Zone 1 predictions DNN'] = predictions
```
